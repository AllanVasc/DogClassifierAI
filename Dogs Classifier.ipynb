# The Incredible Dog Classifier!

This application gives you the power to classify dog ​​breeds in Chow Chow, Samoyed or Labrador. Put a picture of one of these races and watch the magic happen!

This code was developed based on the fastAi classes.

Author: Allan Soares Vasconcelos



```python
path = Path()
path.ls(file_exts='.pkl')
```

```python
path = Path()
learn_inf = load_learner(path/'dogclassifier.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])
```

```python
display(VBox([widgets.Label('Select your dog pic!'), btn_upload, out_pl, lbl_pred]))
```
